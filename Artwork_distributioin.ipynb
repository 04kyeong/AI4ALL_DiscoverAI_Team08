```python
# Set X variable
X = [ 
     1800,
     1825,
     1850,
     1875,
     1900,
     1925,
     1950,
     1975,
     2000
     ]
```

```python
countries = {}

for name in df['Nationality'].unique() :
    
    countries[name] = [0,0,0,0,0,0,0,0,0]
```

```python
for index,row in df.iterrows() :
    name = row['Nationality']
    year = row['Date']
    
    for i in range(len(X)):
        if year >= X[i] and (len(X) == i+1 or year < X[i+1]) :
            countries[name][i] = countries[name][i] + 1
            #print(year,i,X[i])
```

```python
# Set Y varaible
for name in df['Nationality'].unique():
    Y = countries[name]
    for i in Y:
        if (i > 300):
            plt.plot(X,Y, 'o-', label = name)
            break
        else :
            plt.plot(X,Y, 'o-', label = '_nolegend_')
    plt.legend(loc = 'upper left')
    plt.title('Amount of Artworks from each country though the decades')
    plt.xlabel('Year')
    plt.ylabel('Number of Artworks')
```

```python
for name in df['Nationality'].unique():
      fig, ax = plt.subplots(figsize=(20,8))
      Y = countries[name]
      ax.plot(X,Y,'o-',label = name)
      ax.legend(loc = 'upper left')
```
